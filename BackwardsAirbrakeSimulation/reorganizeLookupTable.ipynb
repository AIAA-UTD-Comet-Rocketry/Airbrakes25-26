```python
import pandas as pd

table = pd.read_csv("LookupTable.csv", header=None)

newTable = pd.DataFrame()

numAngles = 6

startingIndex = 0

numRows = table.shape[0]
print("NUm rows: " + str(numRows))

for angle in range (0, numAngles, 1):
    for rowIndex in range(angle, numRows, numAngles):
        row = table.iloc[[rowIndex]]
        print(row)
        newTable = pd.concat([newTable, row], ignore_index=True)

newTable.to_csv("FinalLookupTable.csv", header=None, index=False)
```
